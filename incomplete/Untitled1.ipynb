```python
from MeCab import Tagger

mc = Mecab('mecab-ko-dic-2.1.1-20180720')
sentence = '응용유기공업화학이랑 공업수학 이정규교수님 알려줘!'
words = mc.pos(sentence)

print(words)

tagger = Tagger('-d %s' % 'mecab-ko-dic-2.1.1-20180720')

for word in words:
    temp = tagger.parse(word[0])
    #print('temp\t' + temp)
    p = temp.split('\t')[0] # 잘려진 단어
    #print(p)
    n = temp.split('\t')[-1].split(',')[1]
    #print('의미\t' + n)
    if n == '과목명':
        print('찾고자하는 과목은 \'' + p + '\' 입니다.')
```
